```python
"""
================================================================================
【Project IQD：宏觀社會經濟拓樸演化 量子因果引擎核心代碼】
================================================================================
[1. 核心映射表 (Mapping Table)]
▶ Qubits (觀測維度): q0 到 q5 對應易經六爻。
  - q0 (初爻): 社會基層、民生與實體的物理基礎 (如原物料)
  - q2 (三爻): 供應鏈中段的製造業與全球物流樞紐 (Hub)
  - q5 (上爻): 大國政府與最高決策體制
▶ Gates (邏輯閘與事件):
  - 初始態 |000000>: 坤為地卦 (絕對靜止、毫無波瀾、還沒有任何事件發生的初始宇宙)
  - Rx 閘 (Rotation): 接收 Theta 數值，將天地煞氣直接注入平靜宇宙，對安穩節點施加偏轉壓力
  - CNOT 閘 (CX): 強制的因果綁定 (量子糾纏)。當控制端變化，目標端必然翻轉 (100%強制力)
▶ 測量輸出 (Bitstring -> 卦象顯化):
  - 010101: 火水未濟卦 (陰陽交錯、動態牽制、局部尋求生路，尚未走到盡頭)
  - 111100: 雷天大壯卦 (陽氣過盛、底層超載著火，頂層無力回天的恐怖系統性傳染)
  - 000100: 地山謙卦   (超級樞紐 q2 單獨承受 1 的極限高壓，形成單點故障的絕對脆弱孤立之象)
[2. 最小案例流程圖 (Minimal Case Flow)]
輸入歷史客觀壓力 ──> [數值轉換] 映射為具體物理張力角度 (Theta) 
                 ──> [張力注入] 系統由坤為地(0)開局，對指定 Qubit 施加 Rx 閘 (波函數偏轉) 
                 ──> [交感綁定] 依據年代 (1939線性/2008全連接/2024樞紐) 施加 CNOT 閘建立網路結構
                 ──> [最終測量] 波函數顯化 (Collapse) 
                 ──> [結果輸出] 4096 次平行宇宙射擊後，統計出最高機率之主態字串 (映射回現實卦象)
================================================================================
"""
# ==========================================================
# 步驟 0: 安裝量子計算環境
# ==========================================================
# 確保本機端已安裝套件: pip install qiskit qiskit-aer matplotlib
# ==========================================================
# 步驟 1: 匯入必要的量子函式庫與視覺化補丁
# ==========================================================
import numpy as np
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt
# --- [視覺化優化補丁] ---
# 確保圖表可以顯示中文，且 X 軸標籤不會重疊
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei', 'PingFang TC', 'Arial Unicode MS']
plt.rcParams['axes.unicode_minus'] = False
# ------------------------
# ==========================================================
# 步驟 2: 建構「易經六爻」宏觀演化量子場 (AI 4.0 Model)
# ==========================================================
class QuantumSupplyChain:
    def __init__(self, year, metric_val, metric_type, topology):
        """
        初始化供應鏈環境
        :param year: 模擬年份 (1939, 2008, 2024)
        :param metric_val: 歷史數據值 (關稅%, VIX恐慌指數, GSCPI斷鏈指數)
        :param metric_type: 數據類型 ('percentage', 'index', 'std_dev')
        :param topology: 網絡交感拓樸 ('linear', 'full', 'hub_spoke')
        """
        self.year = year
        self.n_qubits = 6  # 觀測維度：對應 Hexagram 易經六爻 (q0~q5)
        self.topology = topology
        
        # 初始參數：在量子線路剛通電的那一瞬間，初始狀態為絕對的 |000000>。
        # 這對應易經的「坤為地」卦，代表絕對靜止、沒有任何混沌雜訊的初始宇宙。
        self.circuit = QuantumCircuit(self.n_qubits, self.n_qubits)
        
        # 將大自然的歷史壓力轉化為物理張力
        self.theta = self._calibrate_theta(metric_val, metric_type)
    def _calibrate_theta(self, val, type):
        """
        A. 數學原理與校準 (Calibration) - 能量映射
        將真實世界的宏觀指標精準映射為量子旋轉角 Theta [0, Pi]。
        Theta 越大，代表輸入系統的天地煞氣越重，Rx 閘將安穩狀態往風險態偏轉的幅度越大。
        """
        if type == 'percentage':
            # 1939案例：斯姆特霍利關稅法案約 60% 峰值 (假設 100% 關稅 = Pi 完全阻斷)
            return np.pi * (val / 100.0)
        elif type == 'index':
            # 2008案例：VIX 恐慌指數逼近 80 的極端值 (假設 VIX=100 為系統最大張力狀態 Pi)
            return np.pi * (val / 100.0)
        elif type == 'std_dev':
            # 2024案例：GSCPI 供應鏈壓力指數正 4.3 個標準差的離群值
            # 使用 Sigmoid 函數將 (-inf, inf) 非線性映射到 (0, Pi)
            sigmoid = 1 / (1 + np.exp(-(val - 0))) 
            return np.pi * (0.1 + 0.8 * sigmoid)
        return 0.0
    def apply_tension_hamiltonian(self):
        """
        B. 注入「外部張力」 (Tension Injection) - 施加 Rx 閘
        物理意義：系統沒有預設的高熵混沌，直接由 Rx 閘負責接收 Theta 數值，
        模擬外部宏觀壓力無差別打擊，對原本安穩的節點施加強烈的偏轉壓力。
        """
        for i in range(self.n_qubits):
            current_theta = self.theta
            if self.year == 2024 and i in [0, 2]: # 假設 q0 (民生), q2 (物流/半導體) 是受制裁節點
                current_theta *= 1.2 # 額外疊加 20% 的重壓數值
            self.circuit.rx(current_theta, i)
            
        self.circuit.barrier() # 視覺分隔：張力注入階段完成
    def apply_entanglement_topology(self):
        """
        C. 建立「量子糾纏」 (Entanglement) - 構建網絡拓樸與六爻交感
        物理意義：最關鍵的 CNOT 閘，數值為絕對的百分之百強制翻轉。
        因此，決定最終顯化長相的不是交感的強弱，而是「拓樸連線的密度」。
        """
        if self.topology == 'linear':
            # 1939: 線性拓樸。連線極度稀疏，壓力只能單向且緩慢傳導 (q0 -> q1 -> q2...)
            for i in range(self.n_qubits - 1):
                self.circuit.cx(i, i+1)
        elif self.topology == 'full':
            # 2008: 全連接拓樸。極度密集的矩陣，每一個節點都互相連線，形成致命閉環
            for i in range(self.n_qubits - 1):
                self.circuit.cx(i, i+1)
            self.circuit.cx(0, 3)
            self.circuit.cx(1, 4)
            self.circuit.cx(2, 5)
            self.circuit.cx(5, 0) # 閉環反饋點
        elif self.topology == 'hub_spoke':
            # 2024: 樞紐拓樸。所有連線全數壓在 q2 這個超級樞紐上，呈現星狀輻射連線
            hub = 2
            spokes = [0, 1, 3, 4, 5]
            for s in spokes:
                self.circuit.cx(hub, s)
        self.circuit.barrier() # 視覺分隔：量子糾纏結構建立完成
    def measure_hexagram(self):
        """
        D. 觀測量測 (Measurement) - 事件顯化
        將所有互相糾纏的量子態進行測量，瞬間收斂成六個位元的二進位字串(對應易經卦象)。
        """
        self.circuit.measure(range(self.n_qubits), range(self.n_qubits))
    def run(self, shots=4096):
        """執行平行宇宙模擬 (超級電腦射擊 4096 次，讓波函數反覆顯化)"""
        simulator = AerSimulator()
        compiled_circuit = transpile(self.circuit, simulator)
        result = simulator.run(compiled_circuit, shots=shots).result()
        return result.get_counts()

# ==========================================================
# 步驟 3: 運行模擬 (Quantum Simulation)
# ==========================================================
print("✅ 量子環境啟動完成！正在啟動量子供應鏈因果引擎...")
# Case 1: 1939 斯姆特霍利關稅法案 (線性拓樸)
# 預期顯化: 010101 未濟卦 (壓力傳遞緩慢，未形成一致毀滅共識，局部尋求生路)
sim_1939 = QuantumSupplyChain(1939, 60, 'percentage', 'linear')
sim_1939.apply_tension_hamiltonian()
sim_1939.apply_entanglement_topology()
sim_1939.measure_hexagram()
counts_1939 = sim_1939.run()
# Case 2: 2008 金融海嘯 (全連接拓樸)
# 預期顯化: 111100 雷天大壯卦 (系統性傳染引發能量斷層，底層全面著火，頂層無力回天)
sim_2008 = QuantumSupplyChain(2008, 80, 'index', 'full')
sim_2008.apply_tension_hamiltonian()
sim_2008.apply_entanglement_topology()
sim_2008.measure_hexagram()
counts_2008 = sim_2008.run()
# Case 3: 2024 科技戰與斷鏈格局 (樞紐拓樸)
# 預期顯化: 000100 地山謙卦 (結構絕對脆弱，超級樞紐 q2 單獨承受極限高壓的孤立之象)
sim_2024 = QuantumSupplyChain(2024, 4.3, 'std_dev', 'hub_spoke')
sim_2024.apply_tension_hamiltonian()
sim_2024.apply_entanglement_topology()
sim_2024.measure_hexagram()
counts_2024 = sim_2024.run()
# ==========================================================
# 步驟 4: 視覺化成果 (Visualization)
# ==========================================================
print("\n🔍 模擬完成！正在生成全像張力分佈圖...")
# 生成並排的三張歷史拓樸演化圖
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
plot_info = [
    (counts_1939, "1939: 貿易保護主義 Trade Protectionism\n(線性拓樸 Linear)", axes[0]),
    (counts_2008, "2008: 金融海嘯 Financial Tsunami\n(全連接拓樸 Full)", axes[1]),
    (counts_2024, "2024: 科技戰與斷鏈 Technology War and Supply Chain Disruption\n(樞紐拓樸 Hub-Spoke)", axes[2])
]
for counts, title, ax in plot_info:
    # 🟢 降噪處理：過濾掉發生機率極低的量子雜訊，只取 Top 15 相變主態卦象
    top_counts = dict(sorted(counts.items(), key=lambda item: item[1], reverse=True)[:15])
    
    # 繪製機率直方圖 (代表平行宇宙顯化後的共識強度)
    plot_histogram(top_counts, ax=ax)
    ax.set_title(title, fontsize=14, fontweight='bold')
    
    # 🟢 視覺優化：X 軸標籤旋轉 90 度避免重疊
    ax.tick_params(axis='x', labelrotation=90, labelsize=9)
# 自動調整排版並顯示圖表
plt.tight_layout()
```
