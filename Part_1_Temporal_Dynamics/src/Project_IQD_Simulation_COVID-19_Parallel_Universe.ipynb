```python
"""
================================================================================
【Project IQD：COVID-19 初期純混沌實驗 量子因果引擎核心代碼】
================================================================================
[1. 核心映射表 (Mapping Table)]
▶ Qubits (觀測維度): 社會運作骨幹 (6 Qubits，重點觀測其中 4 個)
  - q0: 基層民生與恐慌 (Livelihood/Panic)
  - q2: 底層供應鏈 (Supply Chain)
  - q3: 中堅層/中產階級 (Middle Class)
  - q5: 政府體制與政策 (Government)

▶ Gates (邏輯閘與事件物理意義):
  - H 閘 (Hadamard): 絕對混沌製造機。將所有節點打入不知所措的絕對疊加態 (高熵紊流)。
  - Rx 閘 (Phase Rotation): 局部劇烈震盪。注入外部張力 (pi/2.5)，精準打擊民生與決策節點。
  - CNOT 閘 (CX): 脆弱的交叉影響。沒有死鎖，僅剩盲目的傳染與政策衝撞。

▶ 測量輸出 (Bitstring -> 卦象顯化):
  - 101101 (1.75%): 離為火卦 (疫情與恐慌如火勢般連鎖延燒，缺乏方向)
  - 000100 (1.75%): 地山謙卦 (單一樞紐面臨停擺的極限高壓)
  - 111010 (1.74%): 天水訟卦 (資源匱乏導致社會各階層爆發爭端與衝突)
  - (因為 H 閘的極端干擾，無論 Rx 如何施壓，所有 64 卦的顯化機率皆被徹底攤平至 ~1.5% 左右的量子雜訊)

[2. 最小案例流程圖 (Minimal Case Flow) - 高溫紊流實驗]
[極致混沌] 對所有 Qubit 施加 H 閘 (創造絕對未知的疊加態) 
   ──> [局部震盪] 設定張力 theta (pi/2.5)，對民生(q0)與決策(q5)施加 Rx 閘
   ──> [脆弱傳染] 施加 CNOT 閘 (政府 q5 盲目牽動供應鏈 q2)
   ──> [恐慌蔓延] 施加 CNOT 閘 (民生 q0 恐慌傳染給中堅層 q3)
   ──> [最終測量] 波函數顯化 (Collapse)
   ──> [結果輸出] 拉滿算力執行 20000 次射擊，觀測徹底失序的機率攤平現象
================================================================================
"""

# ==========================================================
# 步驟 0: 安裝量子計算環境 (只需 10 秒)
# ==========================================================
# !pip install qiskit qiskit-aer pylatexenc > /dev/null
print("✅ 量子環境安裝完成！正在啟動 AI 4.0 模擬...")

# ==========================================================
# 步驟 1: 匯入必要的量子函式庫
# ==========================================================
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt
import numpy as np

# --- [新增] 修復 Matplotlib 中文顯示 (豆腐塊) 問題 ---
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei', 'PingFang TC', 'Arial Unicode MS'] # 依序支援 Win, Mac
plt.rcParams['axes.unicode_minus'] = False # 確保負號不會變成方塊
# ------------------------------------------------------

# ==========================================================
# 步驟 2: 建構「易經六爻」量子場 (AI 4.0 Model)
# ==========================================================
# 初始化 6-Qubit 系統 (對應六爻)
qc = QuantumCircuit(6)

# A. 創造疊加態 (Superposition) - 混沌未開
# 物理意義：疫情初期，所有節點(0~5)皆處於不知所措的高熵絕對疊加態
qc.h(range(6))

# B. 注入「外部張力」 (Tension Injection) - 模擬 COVID-19 壓力
# 物理意義：對初爻(民生)和五爻(政府)施加強烈的相位旋轉 (Phase Rotation)，模擬局部震盪
tension = np.pi / 2.5
qc.rx(tension, 0) # 民生震盪
qc.rx(tension, 5) # 決策震盪

# C. 建立「量子糾纏」 (Entanglement) - 全球化連動
# 物理意義：移除了剛硬的死鎖，只保留最基本的隨機連動與盲目傳染
# 政府(Q5)的決策會強制改變供應鏈(Q2)的狀態
qc.cx(5, 2)
# 民生(Q0)的恐慌會傳導到社會中堅層(Q3)
qc.cx(0, 3)

# D. 觀測量測 (Measurement) - 事件顯化
qc.measure_all()

# ==========================================================
# 步驟 3: 運行模擬 (Quantum Simulation)
# ==========================================================
print("🚀 算力拉滿！啟動 20000 次平行宇宙射擊測試波函數無序性...")
simulator = AerSimulator()
compiled_circuit = transpile(qc, simulator)
job = simulator.run(compiled_circuit, shots=20000) # 模擬 20,000 個平行宇宙
result = job.result()
counts = result.get_counts(qc)

# ==========================================================
# 步驟 4: 視覺化成果 (Visualization)
# ==========================================================
print("\n🔍 模擬結果分析 (前 5 大可能發生的相變卦象):")

# 對所有 64 種可能顯化的卦象進行排序
sorted_counts = sorted(counts.items(), key=lambda x: x[1], reverse=True)

# 顯示前五大機率卦象 (證明機率被攤平至 ~1.75% 附近)
for hex_code, count in sorted_counts[:5]:
    probability = (count / 20000) * 100
    # 將二進位轉為簡單解釋
    print(f"卦象代碼 [ {hex_code} ] -> 出現機率: {probability:.2f}%")

print("\n(註：所有機率皆被攤平，證明在絕對疊加態的干擾下，局部張力無法形成共識，未來無法預測。)")
print("\n正在繪製張力分佈圖...")

# 繪製機率直方圖 (展示這片「平坦的荒原」)
plot_histogram(counts, figsize=(20, 6), title="AI 4.0 Prediction: 新冠疫情社會張力相變 COVID-19 Social Phase Transition under Tension")
```
