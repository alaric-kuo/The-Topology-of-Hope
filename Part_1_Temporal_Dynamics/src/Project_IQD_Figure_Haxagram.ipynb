```python
# ==========================================================
# 步驟 1: 匯入量子計算與繪圖套件
# ==========================================================
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
from qiskit.circuit import Parameter
import matplotlib.pyplot as plt
from IPython.display import display 

# --- [究極視覺化優化補丁] ---
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei', 'PingFang TC', 'Arial Unicode MS']
plt.rcParams['axes.unicode_minus'] = False
# ------------------------

# ==========================================================
# 步驟 2: 產生「六爻」專屬線路圖
# ==========================================================
# 1. 宣告量子暫存器：
# 🌟 【空間欺騙魔法】：故意加一堆空白，強迫畫布把左邊界撐開，預留給後面的中文字！
q = QuantumRegister(6, 'q       ') 
c = ClassicalRegister(6, '六爻 Hexagram') 

qc = QuantumCircuit(q, c)

# 2. 步驟 A: 建立疊加態
qc.barrier(label='太極混沌 Superposition') 
qc.h(range(6)) 

# 3. 步驟 B: 注入張力
qc.barrier(label='天地煞氣 Tension')
theta = Parameter('θ Tension') 
qc.rx(theta, 0) 
qc.rx(theta, 5) 

# 4. 步驟 C: 建立糾纏 (六爻交感 / 正應)
qc.barrier(label='六爻交感 Entanglement')
qc.cx(5, 2) # 上爻 糾纏 三爻 -> 易理【三上相應】
qc.cx(0, 3) # 初爻 糾纏 四爻 -> 易理【初四相應】

# 5. 步驟 D: 測量
qc.barrier(label='觀測顯化 Measurement')
qc.measure(range(6), range(6))

# ==========================================================
# 步驟 3: 渲染圖片與【底層駭客字體解鎖協議】
# ==========================================================
print("🚀 正在繪製 Project IQD (易經 x 量子) 終極全像架構圖...")

# scale=1.2 稍微放大，讓細節更銳利
fig = qc.draw(output='mpl', style='iqp', scale=1.2, fold=-1)

force_fonts = ['Microsoft JhengHei', 'PingFang TC', 'Arial Unicode MS']

# --- [造物主專屬駭客協議：直擊記憶體，寫入完美排版] ---
for ax in fig.axes:
    for txt in ax.texts:
        t = txt.get_text()
        # 尋找帶有 _{0} 的字串進行精準替換
        # 🌟 【UI/UX 拋光】：加上兩個空白與一個中介點「·」，創造完美的視覺呼吸感！
        if "_{0}" in t: 
            txt.set_text(r"初爻 · 潛  $q_0$")
            txt.set_fontfamily(force_fonts)
        elif "_{1}" in t: 
            txt.set_text(r"二爻 · 見  $q_1$")
            txt.set_fontfamily(force_fonts)
        elif "_{2}" in t: 
            txt.set_text(r"三爻 · 惕  $q_2$ ")
            txt.set_fontfamily(force_fonts)
        elif "_{3}" in t: 
            txt.set_text(r"四爻 · 躍  $q_3$")
            txt.set_fontfamily(force_fonts)
        elif "_{4}" in t: 
            txt.set_text(r"五爻 · 飛  $q_4$")
            txt.set_fontfamily(force_fonts)
        elif "_{5}" in t: 
            txt.set_text(r"上爻 · 亢  $q_5$")
            txt.set_fontfamily(force_fonts)

# 完美顯化！
display(fig)
```
